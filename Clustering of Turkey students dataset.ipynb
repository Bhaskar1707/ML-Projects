# Lets begin to cluster the students based on the  questionnaire  data

```python
df_questions = df.iloc[:,5:33]
```

```python
df_questions.head()
```

# Lets do a PCA for feature dimensional reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
df_questions_pca = pca.fit_transform(df_questions)
```

```python
from sklearn.cluster import KMeans
SSE = []
for i in range(1, 7):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_questions_pca)
    SSE.append(kmeans.inertia_)
plt.plot(range(1, 7), SSE)
plt.title('Elbow Method')
plt.xlabel('Number of clusters(k)')
plt.ylabel('SSE')
plt.show()
```

# Based on the above Elbow Method graph , we can choose for 3 clusters for this dataset.

# Now, visulaizing the clusters with their centroids-

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(df_questions_pca)
```

```python
plt.scatter(df_questions_pca[y_kmeans == 0, 0], df_questions_pca[y_kmeans == 0, 1], s = 100, c = 'orange', label = 'Cluster 1')
plt.scatter(df_questions_pca[y_kmeans == 1, 0], df_questions_pca[y_kmeans == 1, 1], s = 100, c = 'red', label = 'Cluster 2')
plt.scatter(df_questions_pca[y_kmeans == 2, 0], df_questions_pca[y_kmeans == 2, 1], s = 100, c = 'black', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c = 'silver', label = 'Centroids',s= 300)
plt.title('Clusters of students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
